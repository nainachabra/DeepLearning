```python
#IMPORTING  LIBRARIES
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
```

READING THE DATASET

```python
df= pd.read_csv("bodyfat.csv")
df
```

```python
df.shape
```

NOTING DOWN THE NUMBER OF SAMPLES FEATURES AND OUTPUT CLASSES

```python
num_samples = df.shape[0]
num_features = df.shape[1]
num_classes = 1
```

```python
y = df['BodyFat']
X = df.drop('BodyFat',axis=1)       #REMOVING THE DEPENDENT VARIABLE
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)       #SPLITTING THE DATASET INTO TRAINING AND TESTING
```

```python
# KERAS SEQUENTIAL MODEL WITH 3 HIDDEN LAYERS OUT OF WHICH THE FIRST ONE HAS 128 NEURONS THE SECOND HAS 64 AND THE THIRD HAS 32 NEURONS WITH THE OUTPUT NEURON HAVING ONLY ONE OUTPUT NEURON

model = Sequential([
    Input(shape=(num_features-1,),name="s"),
    Dense(128,activation="relu",name="d1"),
    Dense(64,activation="relu",name="d2"),
    Dense(32,activation="relu",name="d3"),
    Dense(1,activation="linear")
])
```

```python
model.summary()         #SUMMARIZING
```

```python
#COMPILING THE MODEL WITH ADAM ALGORITHM AND SINCE THIS IS A REGRESSION PROBLEM THIS METRICS USED IS ROOTMEANSQUAREDERROR AND LOSS IS MEAN SQUEARED LOSS

model.compile(optimizer="adam",loss="mean_squared_error",metrics=['RootMeanSquaredError'])
```

```python
#FITING THE MODEL WITH TRAINING DATA BY TAKING THE BATCH SIZE AS 32 AND 100 EPOCHS ALONG WITH 0.1 VALIDATION SPLIT

model.fit(X_train,y_train,batch_size=32,epochs=100,validation_split=0.1)
```

```python
# MODEL BASED ON THE TEST DATA AND BATCH SIZE AS 32

loss,accuracy = model.evaluate(X_test,y_test,batch_size=32)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
